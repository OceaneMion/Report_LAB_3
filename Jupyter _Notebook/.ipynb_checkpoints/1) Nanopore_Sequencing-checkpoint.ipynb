# Data and focus of the study

* Our project focus on identifying modified bases (5mC: methylation, 5hmC:demethylation) in *Pseudogymnoascus destructans* variants and determine if there is an association with the genomic position of transposable elements (TEs).

* For this purpose, we have 11 variants from 2 lineages (1 and 2), as well as a supposed non pathogenic closely related outgroup (*Pseudogymnoascus spp*) on which was performed nanopore sequencing

* Nanopore sequencing was performed on the conidium (spores) of the fungus

# What is Nanopore Sequencing and his advantages ?


Nanopore Sequencing is a long read sequencing technique. Compared to short reads sequencing it allow to span more easily repeat and complex element such as transposable elements (TEs) and are therefore of interest for our study.
Furthermore, it allows to differenciation active demethylation marks (5hmC) from traditionnal methylation mark (5mc) which is not the case of traditionnal bisulfite sequencing. As we want to study the association between TEs of *P.destructans* genome and methylation, Nanopore sequencing seems like a suitable technique.

#How does Nanopore sequencing works ?

In this study, we used a MinION sequencer, which contains a flow cell composed of numerous pores. Each pore contains, nanopore protein embedded in an electrically resistant polymer membrane, and immersed in an electrolyte solution. An electrical voltage is applied across the membrane, causing an ionic current to flow through the nanopores. Double stranded (dsDNA) is introduced to the negatively charged 'cis' side of the membrane. A motor protein, typically a helicase, unwinds the dsDNA into single strand DNA (ssDNA). The negatively charged DNA is then drawn through the nanopore to the positively charged 'trans' side due to the applied voltage. As each nucleotide passes through the nanopore, it causes a characteristic disruption in the ionic current, specific to the type of nucleotide (A, T, C, G), and presence of methylation marks. These current disruptions are detected by a sensor connected to each nanopore, producing a signal often referred to as a 'squiggle.' Specialized algorithms analyze these disruptions in real-time to determine the sequence of bases passing through the pore. This process, known as basecalling (See Basecalling Notebook), uses machine learning methods based on neural networks trained on large datasets (Bacteria + Human), allowing for accurate reconstruction of the DNA sequence.
